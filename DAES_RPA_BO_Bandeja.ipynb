```python
# Instala Chrome y ChromeDriver en el entorno de Colab (Linux)
!apt update
!apt install -y chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin


# Instalar Kora para hacer compatibles chrome con chromedrivers
!pip install kora

# Instalar Selenium y WebDriver Manager
!pip install selenium webdriver-manager


# Importar librerías necesarias
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager


import time
# Import the Service class
from selenium.webdriver.chrome.service import Service

# Configurar opciones para Chrome
options = webdriver.ChromeOptions()
#options.add_argument("--headless")  # Modo sin interfaz gráfica
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
# Add arguments to resolve the DevToolsActivePort issue
options.add_argument("--remote-debugging-port=9222")  # Use a specific port for debugging
options.add_argument("--disable-gpu") # Disable GPU acceleration, sometimes this can cause issues
options.add_argument("--disable-extensions") #Disable extensions, which can interfere

# Add the following line to specify the path to the Chrome binary
# Initialize the ChromeDriverManager to automatically download and manage the driver
chrome_manager = ChromeDriverManager()
driver_path = chrome_manager.install()
#options.binary_location = driver_path  # Set the binary location using the manager
#options.binary_location = "/usr/bin/chromium-browser" # Explicitly tell it where Chrome is
options.binary_location = chrome_manager.driver.binary_path # Tell it to use the path from ChromeDriverManager


# Print chromedriver and chrome versions
#print(f"Chromedriver version: {chrome_manager.driver.version}") # print the version
!chromium-browser --version # print the chrome version


# Inicializar el navegador con la versión correcta de ChromeDriver
driver = webdriver.Chrome(service=Service(driver_path), options=options) # this uses the path we downloaded

try:
    # Paso 1: Ir a la página de autenticación
    driver.get("https://backofficeaes.economia.cl/Account/Login")

    # Esperar a que el campo de usuario esté presente
    wait = WebDriverWait(driver, 10)
    user_field = wait.until(EC.presence_of_element_located((By.NAME, "UserName")))
    password_field = driver.find_element(By.NAME, "Password")
    login_button = driver.find_element(By.ID, "btnLogin")

    # Paso 2: Ingresar usuario y contraseña
    user_field.send_keys("lpachecoadmin")
    password_field.send_keys("lpachecoadmin$")

    # Paso 3: Iniciar sesión
    login_button.click()

    # Esperar a que cargue la página principal
    wait.until(EC.url_contains("backofficeaes.economia.cl"))

    # Verificar la URL después del inicio de sesión
    print("URL actual después de autenticación:", driver.current_url)

    # Paso 4: Ir a la bandeja de tareas
    driver.get("https://backofficeaes.economia.cl/Inbox")
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))  # Esperar a que cargue la página

    # Paso 5: Usar el tabulador 25 veces para enfocar el campo de filtro
    body = driver.find_element(By.TAG_NAME, "body")
    for _ in range(25):
        body.send_keys(Keys.TAB)
        time.sleep(0.1)

    # Paso 6: Escribir "AREA ASIGNACIÓN" en el campo de filtro
    body.send_keys("AREA ASIGNACIÓN")
    time.sleep(3)

    print("Automatización completada.")

except Exception as e:
    print(f"Ocurrió un error: {e}")

finally:
    # Cerrar el navegador
    driver.quit()
```
